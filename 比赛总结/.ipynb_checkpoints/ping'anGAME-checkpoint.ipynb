# 数据处理工具的使用

# pandas

> 这是一个可以用来读取与处理csv文件（目前只涉及到了这个功能）

```python
import pandas as pd
import numpy as np
```

##### pd.read_csv获得的是一个dataframe对象。该对象是一个“表格”对象，可以直接输出，空白处为NAN值

```python
data_frame = pd.read_csv("../datas/csvTest/readTest.csv")  #  ../是上级目录
```

```python
data_frame
```

##### 获取列

```python
print(data_frame["member_id"] )#通过列名索引
data_frame["loan_amnt"][1] #获取单个元素
```

```python
print(data_frame[["member_id","loan_amnt"]]) #获取多列,注意不能用数字索引哦
```

##### 获取行（注意：loc是索引在读取时的左侧的编号，iloc是索引绝对编号（行数））

```python
test = data_frame[1:3]
print(test) #通过数字索引行
print("----------------------------------------------------------------")
print(data_frame[:0]) #首先不能直接通过数字索引，其次用数字时第一行（名字那行）是要计数的
```

```python
print(test["member_id"]) #得到的还是一个dataframe
```

```python
test = data_frame.loc[1] #获取出第一行的各值
print(test) 
print(test[1])
```

```python
test = data_frame.loc[1:3] #取出多行,可以对比一下直接索引，这是不计第一行（名字那行）的
print(test)
```

```python
test = data_frame.loc[[1,4]] #通过行名索引出不连续的行
print(test)
```

```python
test = data_frame.iloc[1]
print(test)
```

```python
print(data_frame.at[5,"member_id"])  #直接通过“坐标位置”进行索引
print(data_frame.iat[5,0])  #直接通过纯粹的坐标进行索引
```

##### 填补NAN值项

```python
print(data_frame["emp_title"][7])
```

```python
data_frame = data_frame.fillna(1)
print(data_frame["emp_title"][7])
```

##### 删除

```python
data_frame = data_frame.drop([6,8],axis=0) #删除第6行和第8行，维度为0
print(data_frame)
```

##### 插入数据

```python
data_frame.insert(0,"say","Hello")  #第一个为第几列，第二个是名字
data_frame
```

```python
data_frame=data_frame.drop(["say"],axis=1)
```

```python
data_frame
```

```python
temp = data_frame.pop("member_id")#pop
print(temp) 
```

```python
data_frame
```

##### 迭代器

```python
iter = data_frame.iteritems()
for i in data_frame:
    print(i)  #输出所有列|名（第一行）
```

```python
for i in iter:
    print("test",i)  #按列输出所有元素
```

```python
iter = data_frame.iterrows()
for i in iter:
    print("test:",i)  #按行输出
```

```python
iter = data_frame.itertuples()
for i in iter:
    print(i)  #把每行当作一个元组，进行遍历
```

## numpy

```python
#注意(5,)和(1,5)的区别，还有就是[5,1]和[1,5]的区别
array = np.array([1,2,3,4,5],dtype="float")
print(array.shape)
print(array)
array = array.reshape([5,1])
print(array)
array = array.reshape([1,5])
print(array)
print(array.shape)
```

##### 删除

```python
a = np.delete(array,3,axis=1) #删除第三列
```

```python
print(a)
```

```python
y = [temp*2 for temp in a]  #对每个元素操作，然后装入list中
print(y)
```

## random

```python
import random
```

```python
array = array.reshape((5,))
test = random.sample(list(array),3) #从array中随机选取出3个
print(test)
```

## Pickle(对象持久化)

```python
import pickle 
```

```python
test_dict={"wang":1,"tian":2,"rui":3}
f = open("../datas/csvTest/pickletest.txt","wb")
```

```python
pickle.dump(test_dict,f)
f.close()
```

```python
f = open("../datas/csvTest/pickletest.txt","rb")
read_dict = pickle.load(f)
print(read_dict)
```
